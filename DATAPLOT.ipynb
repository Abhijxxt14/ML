```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# -------- Sample Data --------
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 8, 10])

# -------- Train Linear Regression (for line) --------
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# =====================================
# 1 SCATTER PLOT
# =====================================
plt.figure()
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter Plot")
plt.show()

# =====================================
# 2 REGRESSION LINE PLOT
# =====================================
plt.figure()
plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, label="Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Regression Line")
plt.show()

# =====================================
# 3 HISTOGRAM
# =====================================
plt.figure()
plt.hist(y, bins=5)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()

# =====================================
# 4 BOXPLOT
# =====================================
plt.figure()
plt.boxplot(y)
plt.ylabel("Values")
plt.title("Boxplot")
plt.show()
```
