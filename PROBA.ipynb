```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom

# =====================================
# 1 RANDOM SAMPLE GENERATION
# =====================================
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1
print("Random samples generated")

# =====================================
# 2 NORMAL DISTRIBUTION
# =====================================
x = np.linspace(-4, 4, 100)

# PDF (Probability Density Function)
pdf = norm.pdf(x, loc=0, scale=1)

# CDF (Cumulative Distribution Function)
cdf = norm.cdf(x, loc=0, scale=1)

# Plot PDF and CDF
plt.figure()
plt.plot(x, pdf, label="Normal PDF")
plt.plot(x, cdf, label="Normal CDF")
plt.legend()
plt.title("Normal Distribution: PDF & CDF")
plt.show()

# =====================================
# 3 HISTOGRAM vs THEORETICAL PDF
# =====================================
plt.figure()
plt.hist(data, bins=30, density=True, alpha=0.6, label="Histogram")
plt.plot(x, pdf, label="Theoretical PDF")
plt.legend()
plt.title("Histogram vs PDF")
plt.show()
n = 10      # number of trials
p = 0.5     # probability of success
k = np.arange(0, n+1)

binom_pmf = binom.pmf(k, n, p)
binom_cdf = binom.cdf(k, n, p)

# Plot Binomial PMF
plt.figure()
plt.stem(k, binom_pmf, use_line_collection=True)
plt.title("Binomial Distribution (PMF)")
plt.xlabel("k")
plt.ylabel("Probability")
plt.show()
```
