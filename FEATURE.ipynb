```python

import pandas as pd

# -------- Create Sample DataFrame --------
data = {
    'Maths': [70, 80, 90],
    'Physics': [65, 75, 85],
    'Chemistry': [60, 70, 80]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# =====================================
# 1️⃣ CREATE NEW FEATURE
# =====================================
# Total marks feature
df['Total'] = df['Maths'] + df['Physics'] + df['Chemistry']

# Average marks feature
df['Average'] = df['Total'] / 3

print("\nAfter Adding New Features:\n", df)

# =====================================
# 2️⃣ REMOVE COLUMN
# =====================================
df_removed = df.drop('Total', axis=1)
print("\nAfter Removing 'Total' Column:\n", df_removed)

# =====================================
# 3️⃣ FEATURE SELECTION vs EXTRACTION
# =====================================
# Feature Selection: choosing subset of existing features
selected_features = df[['Maths', 'Physics']]

# Feature Extraction: creating new features (e.g., Total, Average)
extracted_features = df[['Average']]

print("\nFeature Selection:\n", selected_features)
print("\nFeature Extraction:\n", extracted_features)
```
