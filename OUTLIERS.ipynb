```python

import numpy as np
import matplotlib.pyplot as plt

# -------- Sample Data --------
data = np.array([10, 12, 14, 15, 18, 19, 20, 22, 100])  # 100 is an outlier

# =====================================
# 1️⃣ QUARTILES
# =====================================
Q1 = np.percentile(data, 25)   # First Quartile
Q3 = np.percentile(data, 75)   # Third Quartile
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

# =====================================
# 2️⃣ OUTLIER BOUNDS
# =====================================
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# =====================================
# 3️⃣ IDENTIFY OUTLIERS
# =====================================
outliers = data[(data < lower_bound) | (data > upper_bound)]
print("Outliers:", outliers)

# =====================================
# 4️⃣ REMOVE OUTLIERS
# =====================================
clean_data = data[(data >= lower_bound) & (data <= upper_bound)]
print("Data after removing outliers:", clean_data)

# =====================================
# 5️⃣ BOXPLOT
# =====================================
plt.boxplot(data)
plt.title("Boxplot (Outlier Detection)")
plt.show()
```
